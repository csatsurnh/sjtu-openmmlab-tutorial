```python
from mmcls.models import ImageClassifier,GlobalAveragePooling, MobileNetV2, LinearClsHead

model = ImageClassifier(
    backbone=dict(type='MobileNetV2', widen_factor=1.0),
    neck=dict(type='GlobalAveragePooling'),
    head=dict(
        type='LinearClsHead',
        num_classes=30,
        in_channels=1280),
)

model.cuda()
model.load_state_dict(torch.load('fruit30_mmcls.pth')["state_dict"])
```

```python
import torch

imset = []
lbset = []

model.eval()

for i in range(0, len(dataset), 10):
    with torch.no_grad():
        img = dataset[0][0]
        # img = img.flip(0)
        img = img.cuda()
        vec = model.forward_test(img.unsqueeze(0))[0].argmax()

        imset.append(vec)
        lbset.append(dataset[i][1])
    
    # break
    
# immat = torch.stack(imset)
# lbset = torch.tensor(lbset)
```
